```python
import pandas as pd

# List of files to process
files = [
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2023_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2022_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2021_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2020_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2019_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2018_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2017_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2016_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2015_raw.xlsx',
    '/Users/rifatkhan/Documents/Netflix_Raw_Data/Financial_Report_2014_raw.xlsx'
]

# Dictionary to store the extracted data
extracted_data = {
    'Year': [], 'Revenue': [], 'Net Income': [], 'Total Equity': [], 'Total Liabilities': []
}

# Helper function to find the best matching sheet name
def find_best_sheet(sheet_names, target):
    exact_match = [name for name in sheet_names if name.replace('_', ' ').lower() == target]
    if exact_match:
        return exact_match[0]
    
    # Fallback to the shortest name containing the target
    partial_matches = [name for name in sheet_names if target in name.replace('_', ' ').lower()]
    return min(partial_matches, key=len) if partial_matches else None

# Loop through each file
for file in files:
    # Load the Excel file
    xls = pd.ExcelFile(file)
    
    # Initialize variables to store data from each relevant sheet
    revenue, net_income, total_equity, total_liabilities = None, None, None, None
    
    # Find and read relevant sheets
    # Look for the best matching sheet for "Consolidated Statements of Operations"
    operations_sheet = find_best_sheet(xls.sheet_names, 'consolidated statements of op')
    balance_sheet = find_best_sheet(xls.sheet_names, 'consolidated balance sheet')
    
    # Extract data from the Consolidated Statements of Operations sheet
    if operations_sheet:
        df = pd.read_excel(xls, sheet_name=operations_sheet)
        
        # Extract Revenue and Net Income (assuming these terms are in the first column)
        revenue_row = df[df.iloc[:, 0].str.contains('Revenue', case=False, na=False)]
        net_income_row = df[df.iloc[:, 0].str.contains('Net Income', case=False, na=False)]
        
        # Extract values (this assumes the second column has the financial values)
        revenue = revenue_row.iloc[0, 1] if not revenue_row.empty else None
        net_income = net_income_row.iloc[0, 1] if not net_income_row.empty else None

    # Extract data from the Balance Sheet sheet
    if balance_sheet:
        df = pd.read_excel(xls, sheet_name=balance_sheet)
        
        # Extract Total Equity and Total Liabilities
        total_equity_row = df[df.iloc[:, 0].str.contains('Total Stockholders', case=False, na=False)]
        total_liabilities_row = df[df.iloc[:, 0].str.contains('Total Liabilities', case=False, na=False)]
        
        # Extract values (assuming the second column has the financial values)
        total_equity = total_equity_row.iloc[0, 1] if not total_equity_row.empty else None
        total_liabilities = total_liabilities_row.iloc[0, 1] if not total_liabilities_row.empty else None
    
    # Append data for the current file
    extracted_data['Year'].append(file.split('_')[-2])  # Extract the quarter from the file name
    extracted_data['Revenue'].append(revenue)
    extracted_data['Net Income'].append(net_income)
    extracted_data['Total Equity'].append(total_equity)
    extracted_data['Total Liabilities'].append(total_liabilities)

# Convert the extracted data into a DataFrame
final_df = pd.DataFrame(extracted_data)

# Save the combined data to a new CSV or Excel file
final_df.to_csv('/Users/rifatkhan/Documents/Netflix_Data/Netflix data.csv', index=False)
print(final_df)
```
